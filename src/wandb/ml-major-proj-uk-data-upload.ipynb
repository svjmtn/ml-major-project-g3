# Separating UK data

```python
uk_owid = owid[owid['iso_code'] == 'GBR'].copy()
uk_ox = ox[(ox['CountryCode'] == 'GBR') & (ox['Jurisdiction'] == 'NAT_TOTAL')].copy()
uk_regionwise_ox =  ox[(ox['CountryCode'] == 'GBR') & (ox['Jurisdiction'] == 'STATE_TOTAL')].copy()
```

# Dropping redundant columns

We can drop a couple of redundant columns at this point (the ones that have the same value for each data point), but we'll keep a few of them like life_expectancy, human_development_index etc although they have the same value so that we can copy them out and drop them later.

```python
uk_ox.head(1)
```

```python
uk_owid.drop(['iso_code', 'continent', 'location'] , inplace=True, axis=1)
```

We already have `ConfirmedCases` and `ConfirmedDeaths` in `uk_owid` so dropping them in `uk_ox` and `uk_regionwise_ox`

```python
uk_ox.drop(['CountryName', 'CountryCode', 'RegionName', 'RegionCode', 'Jurisdiction',
            'ConfirmedCases', 'ConfirmedDeaths'], inplace=True, axis=1)
```

```python
uk_regionwise_ox.drop(['CountryName', 'CountryCode', 'RegionCode', 'Jurisdiction',
                      'ConfirmedCases', 'ConfirmedDeaths'], inplace=True, axis=1)
```

# Joining the datasets

First we need to make the dates the same format.

```python
uk_owid['date'] = pd.to_datetime(uk_owid['date'], format='%Y-%m-%d')
```

```python
uk_ox['Date'] = pd.to_datetime(uk_ox['Date'], format='%Y%m%d')
uk_regionwise_ox['Date'] = pd.to_datetime(uk_regionwise_ox['Date'], format='%Y%m%d')
```

```python
uk_ox.rename(columns={'Date':'date'}, inplace=True)
```

```python
uk_combined = uk_owid.join(uk_ox.set_index('date'), on='date', how='outer').copy()
```

```python
uk_combined.reset_index(inplace=True, drop=True)
uk_regionwise_ox.reset_index(inplace=True, drop=True)
```

```python
uk_combined.head(2)
```

# Saving CSVs

```python
uk_combined.to_csv('uk_combined.csv', index=False)
uk_regionwise_ox.to_csv('uk_regionwise_ox.csv', index=False)
```

```python
new_uk_artifact = wandb.Artifact('covid19-uk-raw-dataset', type='dataset')

# adding all files to the new artifact folder
new_uk_artifact.add_file('uk_combined.csv')
new_uk_artifact.add_file('uk_regionwise_ox.csv')
run.log_artifact(new_uk_artifact)
```
